```python
import pandas as pd
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("input_data_receipt.csv")
```

```python
original_data = data.copy()
original_data = original_data.dropna()
```

```python
original_data['log_inbnd_duration_min'] = np.log1p(original_data['inbnd_duration_min'])
```

```python
plt.figure(figsize=(8, 4))
#      seaborn
sns.scatterplot(data=original_data, x='sku_cnt', y='inbnd_duration_min', alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.title('Seaborn Scatter Plot with Log-Log Scale')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
#      seaborn
sns.scatterplot(data=original_data, x='sku_cnt', y='log_inbnd_duration_min', alpha=0.6)
plt.xscale('log')
plt.yscale('log')
plt.title('Seaborn Scatter Plot with Log-Log Scale')
plt.show()
```
